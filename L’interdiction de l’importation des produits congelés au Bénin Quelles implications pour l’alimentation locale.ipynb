## Analysons les données liées à l'immportation des viandes au Bénin

```python
# Importons la data relative à cette analyse

d1 = pd.read_csv(r"D:\Notions\Projets d'analyse de données\Interdiction importation des produits congelés au Bénin(données en Tonnes)\donnees_importation_viande_tonnes.csv")
d1
```

## Les contratres entre production nationale de viandes et importation nationale de viandes

```python
# Filtrer les données de d3 pour correspondre à la plage de d1 (2016-2021)
d3_filtered = d3[(d3['Année'] >= 2016) & (d3['Année'] <= 2021)]

# Création du graphique à barres empilées
plt.figure(figsize=(12, 8))
plt.bar(d3_filtered['Année'], d3_filtered["Total Production de viande (T)"], label="Quantité de viandes produites", color='red')
plt.bar(d1['Année'], d1["Volume brut d'importations de la viande (tonnes)"], bottom=d3_filtered["Total Production de viande (T)"], label="Quantité de viandes importées", color='blue')

plt.xlabel('Année')
plt.ylabel('Quantité (tonnes)')
plt.title("Différence entre la quantité nationale de viande produite et le volume de viande importée (2016-2021)")
plt.legend()
plt.show()
```

```python
# Si le gouvernement désire interdire complètement l'importation de viandes au Bénin,alors il faudrait travailler à combler ce gap. Ici il faut remarquer que le gap est juste la moyenne de l'import annuel de viande

gap = moyenne_volume_import_viande

# Déterminons aussi le facteur x  qui représente le nombre de fois que la production nationale de viande doit augmenter pour combler le gap.

facteur_x = (gap / moyenne_viande_produite) - 1

# Afficher les résultats
print(f"Gap à combler (Moyenne des importations annuelles) : {gap:.2f} tonnes")
print(f"Valeur x (facteur de multiplication nécessaire) : {facteur_x + 1:.2f}")
print(f"Donc, la production nationale de viandes au Bénin devra être augmentée de {facteur_x*100:.2f}% pour combler le gap des importations moyennes annuelles, ce qui correspond à multiplier la production actuelle par {facteur_x + 1:.2f}.")
```
